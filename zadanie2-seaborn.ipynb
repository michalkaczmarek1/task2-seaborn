<b>Jakie źródła pozyskania pracownika (Recruitment Source) są najlepsze, jeśli zależy nam na jak najdłuższym stażu pracowników?</b>

```python
df[['DateofTermination','DOB','DateofHire']].dtypes
```

```python
df['DOB'] = pd.to_datetime(df['DOB'],format='%m/%d/%y')
df['DateofTermination'] = pd.to_datetime(df['DateofTermination'],format='%m/%d/%y')
df['DateofHire'] = pd.to_datetime(df['DateofHire'],format='%m/%d/%Y')
```

```python
df[['DateofTermination','DOB','DateofHire']].head()
```

```python
import datetime as dt

def count_seniority(row):

    if pd.isnull(row['DateofTermination']):
        end_date = dt.datetime(2019,9,27)
    else:
        end_date = row['DateofTermination']

    return (end_date - row['DateofHire'])/np.timedelta64(365,'D')
```

```python
df['Seniority'] = df.apply(lambda row: count_seniority(row),axis=1)
```

```python
df[['DateofHire','DateofTermination','Seniority', 'DOB']].head()
```

```python
df['RecruitmentSource']
```

```python
sns.jointplot(x='Seniority',y='RecruitmentSource',data=df)
```

```python
df_recruitment = df.pivot_table(values='Seniority', columns='RecruitmentSource').transpose().sort_values(by='Seniority', ascending=False)
df_recruitment
```

```python
plt.figure(figsize=(55,15))
sns.countplot(x=df['RecruitmentSource'], data=df)
```

Patrząc na wykres, zestawienie (średnia stażu pracy dla danego żródła pozyskania pracownika) oraz odpowiednią reprezentację liczbową pracowników dla danego źródła pozyskania najlepszymi źródłami są: Billboard, MBTA ads, Website Banner Ads, Professional Society.

<b>Jak wygląda struktura wieku aktualnie zatrudnionych pracowników?</b>

```python
from datetime import date

def count_age_employes(row):

    if pd.isnull(row['DateofTermination']):
        end_date = pd.to_datetime(date.today(),format='%m/%d/%y')
    else:
        return 0
    
    
    return round(np.abs((end_date - row['DOB'])/np.timedelta64(365,'D')),0)
```

```python
df['Age_employes'] = df.apply(lambda row: count_age_employes(row),axis=1)
```

```python
plt.figure(figsize=(15,5))
sns.countplot(x=df['Age_employes'][df['Age_employes']>0],data=df)
```

Najwięcej jest osób w wieku 45 lat, następnie kolejno 37 lat oraz 40 lat. Ogólnie najwięcej pracowników jest w przedziale wiekowym 34-45 lat.

<b>Czy starsi pracownicy pracują nad większą liczbą specjalnych projektów niż młodsi pracownicy?</b>

```python
plt.figure(figsize=(15,5))
sns.jointplot(x=df['Age_employes'], y=df['SpecialProjectsCount'], data=df)
```

Starsi pracownicy nie pracują nad większą liczbą specjalnych projektów. Największą liczbę specjalnych projektów robią pracownicy w wieku 34-45 lat wg wykresu.
